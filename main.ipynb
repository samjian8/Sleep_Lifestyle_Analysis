```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns


df = pd.read_csv('data/sleep_health_lifestyle_dataset.csv')

df.head()
```

```python
df.info()
```

```python
df.fillna("None", inplace = True)
```

```python
df.describe()
```

```python
df = df.set_index('Person ID')
df.head()
```

```python
df.dtypes
```

```python
df.duplicated()
```

```python
ax = df['Occupation'].value_counts() \
.head(5) \
.plot(kind='bar', title='Occupation Counts')

ax.set_xlabel('Occupation')
ax.set_ylabel('Count')
```

```python
ax = df["Sleep Duration (hours)"].plot(kind='hist', title='Sleep Durations')
ax.set_xlabel('Sleep Duration in Hours')
```

```python
ax = df["Sleep Duration (hours)"].plot(kind='kde', title='Sleep Durations')
ax.set_xlabel('Sleep Duration in Hours')
```

```python
df.plot(kind='scatter', 
        x='Sleep Duration (hours)',
        y='Quality of Sleep (scale: 1-10)')

# plt.show()
```

```python
sns.scatterplot(x='Sleep Duration (hours)',
        y='Quality of Sleep (scale: 1-10)',
        data=df)
```
